```python
df_activity = pd.read_csv("../DATA/travel_df_final.csv", index_col=0)

df_activity.tail()
```

```python
datas = []
for idx in range(len(df_activity)):
    type2 = df_activity['type2'].loc[idx]
    if  type2 != "0":
        data = "[" + df_activity['type'].loc[idx] + ", " + df_activity['type2'].loc[idx] + "]"
    elif type2 == "0":
        data = "[" + df_activity['type'].loc[idx] + "]" 
    datas.append(data)
```

```python
df_activity["type3"] = datas
```

```python
df_activity[df_activity['type2'] == "0"]
```

```python
df_activity['type'] = df_activity['type3']
```

```python
df_activity_2 = df_activity[['label', 'city_name', 'tour_code', 'tour_name', 'review', 'info', 'corpus', 'target', 'keyword', 'type']]
```

```python
# keywords 내에서 # 삭제
for x in range(0, 70):
    df_activity_2['keyword'][x] = df_activity_2['keyword'][x].replace('#', '')
```

```python
df_activity_2.to_csv("../DATA/travel_df_final.csv")
```
