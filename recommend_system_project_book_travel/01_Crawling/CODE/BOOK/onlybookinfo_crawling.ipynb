```python
ISBNs = list(ISBNs)
```

```python
book_sep_infos = []


for isbn in ISBNs:
    url = "http://www.kyobobook.co.kr/product/detailViewKor.laf?mallGb=KOR&ejkGb=KOR&linkClass=&barcode={}".format(isbn)
    response = requests.get(url)
    dom = BeautifulSoup(response.text, "html.parser")
    
    try:
    

        # book_sommelier_intro = dom.select_one(".box_detail_comment > div").text.strip(),
        book_intro = dom.select_one(".box_detail_article").text.strip(),
        book_ex_recom = dom.select_one(".box_detail_recommend")#.text.replace("\n", "").replace("\r", "").replace("\t", ""),
        book_detail_info = dom.select(".content")[0].text.strip().replace("\r\n", "") #요약
        into_book = dom.select(".content")[1].text.strip().replace("\r", "").replace("\n", "").replace("\t", "").replace("닫기", "") # 책 속으로
        pub_review = dom.select(".content")[3].text.strip().replace("\r", "").replace("\n", "").replace("\t", "").replace("닫기", "") #출판사 서평
    except:
        book_intro = ""
        book_ex_recom = ""
        book_detail_info = ""
        intro_book = ""
        pub_review = ""
        
        
        
    book_sep_infos.append({"ISBN" : isbn,
                           #"book_sommelier_intro":book_sommelier_intro, 
                           "book_intro" :  book_intro,
                           "book_ex_recom" :  book_ex_recom,
                           "book_detail_info" : book_detail_info,
                           "into_book" : into_book,
                           "pub_review" : pub_review})
```

```python
df_book_sep_info = pd.DataFrame(book_sep_infos)
df_book_sep_info
```
