# 3. 상품소개 컬럼에서 특정 문자 제거하는 함수

```python
def del_words(content):
    
    remove_words = ["\r\n\r\n1.", "\r\n", "\r\n\r\n**", "\r\n\r\n", "\r\n3", "-", "★", "■", "•", "↓", "=", "/", "※"]
    for remove_word in remove_words:
        content = content.replace(remove_word, "")
        
    return content
```

```python
data_2["intro"] = data_2["introduction"].apply(del_words)
```
