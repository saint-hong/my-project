```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB  # 다항분포 나이브 베이즈 모델
from sklearn.metrics import accuracy_score  # 정확도 계산
```

```python
dtmvector = CountVectorizer()
X_train_dtm = dtmvector.fit_transform(data['corpus'])
print(X_train_dtm.shape)
```

```python
tfidf_transformer = TfidfTransformer()
tfidfv = tfidf_transformer.fit_transform(X_train_dtm)
print(X_train_dtm.shape)
```

```python
list(data['encoder'])
```

```python
mod = MultinomialNB()
mod.fit(tfidfv, list(data['encoder']))
```

```python
MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)
```

```python
#data[data['title_x'] == '실은 나도 식물이 알고 싶었어']
```

```python
X_test_dtm = dtmvector.transform(data['corpus'])  # 테스트 데이터를 DTM으로 변환
tfidfv_test = tfidf_transformer.transform(X_test_dtm)  # DTM을 TF-IDF 행렬로 변환

predicted = mod.predict(tfidfv_test)  # 테스트 데이터에 대한 예측
```

```python
data['second'] = np.nan
for num in range(0, 126):
    proba = mod.predict_proba(tfidfv[num])
    lab = ['가족', '도시', '도전', '맛집', '문화예술', '셀피', '자연친화', '지식추구', '체험중시']
    d = pd.DataFrame(list(zip(proba[0], lab)), columns=["prob", "label"])
    d = d.sort_values(by='prob', ascending=False).reset_index(drop=True)
    if d['label'][0] == '자연친화':
        data['second'].iloc[num] = d['label'][1]
    else :
        data['second'].iloc[num] = d['label'][0]
```
