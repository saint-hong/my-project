### Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier

est = [('tf', TfidfVectorizer())]
pipe = Pipeline(est)
```

```python
from sklearn.model_selection import train_test_split

X = pipe.fit_transform(data['temp']).toarray()
Y = np.vstack((np.array(data['y1']),np.array(data['y2']),np.array(data['y3']))).T

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=4)
```

```python
X_train
```

```python
forest = RandomForestClassifier(random_state=2, max_features=60, n_estimators=100, max_leaf_nodes=200)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
model = multi_target_forest.fit(X_train, y_train)
predicted = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
