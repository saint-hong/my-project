## 4. 코사인 유사도 메트릭스만들기

### (1) 코사인 유사도 계산 및 메트릭스 만들기

```python
from sklearn.metrics.pairwise import cosine_similarity

co_sim = cosine_similarity(bt, bt)
print(co_sim.shape)
print(co_sim[:2])
```

### (2) 유사도 메트릭스를 내림차순으로 정렬

```python
co_sim_sorted = co_sim.argsort()[:, ::-1]
print(co_sim_sorted[:1])
```

### (3) 코사인 메트릭스 피클로 저장하기

```python
with open('co_sim_sorted.pickle', 'wb') as f:
    pickle.dump(co_sim_sorted, f, pickle.HIGHEST_PROTOCOL)
```

## 5. 책 -> 여행 / 여행 -> 책 반환하는 함수

### (1) 함수

```python
def find_sim_data(df, sorted_ind, code):
    code = df[df['code'] == code]
    code_index = code.index.values
    indexes = sorted_ind[code_index, :]
    #print(code_index)
    #print(indexes)
    
    #책일경우
    if code_index < 200:
        similar_indexes = []
        for idx in indexes[0]:
            # 여행데이터에서 가장 비슷한 것 순으로 정렬
            if idx >= 200:
                similar_indexes.append(idx)
        #print(similar_indexes)
    
    #여행일경우
    else:
        similar_indexes = []
        for idx in indexes[0]:
            # 책데이터에서 가장 비슷한 것 순으로 정렬
            if idx < 200:
                similar_indexes.append(idx)
        #print(similar_indexes)
        
    similar_indexes = np.array(similar_indexes).reshape(-1)
    
    return df.iloc[similar_indexes]
```

### (2) Case1 : 책('종의 기원') -> 여행 

```python
find_sim_data(bt_title, co_sim_sorted, '9788956609959')
```
